```python
sp_char = ['&']

countries = netflix.groupby('country')['country'].count()

for i in range(countries.size):
    
    row_1 = countries.index[i].split(', ')
    
    for k in sp_char:
        row_1 = pd.Series(row_1).apply(lambda x: x.replace(k, ''))
    
    if len(row_1) >1:
        
        for j in row_1:
            if (j in countries.index) and (j!='&'):
                countries.loc[j] += 1
                
            elif (j not in countries.index) and (j!='&'):
                countries.loc[j] = 1
        
        countries[countries.index[i]] = None
                                
    else:
        continue
    
    
countries.dropna(inplace=True)
```

```python
countries = countries.apply(lambda x : int(x)).sort_values()

fig = plt.figure(figsize=(15, 10))
plt.xticks(rotation=90)

sample = countries.sample(50)
sns.lineplot(y=sample.values, x=sample.index)
```

```python
def get_directors(data):
    directors = data.groupby('director')['director'].count()
    return directors.index

directors = get_directors(netflix)
print(directors)
```

```python
def search_by_genre(name, df):
    list_shows = pd.DataFrame(columns=df.columns)
    
    for i in range(len(df.index)):
        if name in df['listed_in'][i]:
            list_shows.loc[df.index[i]] = df.iloc[i]
            
    return list_shows

shows = search_by_genre('Anime', netflix)
shows
```

```python
def search_by_director(name, df):
    list_shows = pd.DataFrame(columns=df.columns)
    
    for i in range(len(df.index)):
        if df['director'][i] == name:
            list_shows.loc[df.index[i]] = df.iloc[i]
            
    return list_shows

shows = search_by_director('Şenol Sönmez', netflix)
shows
```

```python
def search_by_title(name, df):
    list_shows = pd.DataFrame(columns=df.columns)
    
    for i in range(len(df.index)):
        if name in df['title'][i]:
            list_shows.loc[df.index[i]] = df.iloc[i]
            
    return list_shows

shows = search_by_title('Kill', netflix)
shows
```

```python
def search_by_cast(name, df):
    list_shows = pd.DataFrame(columns=df.columns)
    
    for i in range(len(df.index)):
        if name in df['cast'][i]:
            list_shows.loc[df.index[i]] = df.iloc[i]
            
    return list_shows

shows = search_by_cast('Cezmi Baskın', netflix)
shows
```
